# Data Visualization Project Generating Bad Charts

```python
## Imports
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

## Bad Line Chart

```python
badLineNames = ["2016 Q1", "2016 Q2", "2016 Q3", "2016 Q4"]
badLineProduct1 = [240, 300, 280, 400]
badLineProduct2 = [300, 320, 150, 160]
badLineProduct3 = [120, 140, 180, 160]
badLineProduct4 = [380, 400, 450, 500]
```

```python
# Plot a better chart using this data
plt.plot(badLineProduct1, label="Product 1")
plt.plot(badLineProduct2, label="Product 2")
plt.plot(badLineProduct3, label="Product 3")
plt.plot(badLineProduct4, label="Product 4")
plt.xticks(range(len(badLineNames)),badLineNames)
plt.savefig("badLine.png")
```

## Bad Pie Chart

```python
badPieNames = ["Golden", "Boulder", "Denver", "Colo Springs"]
badPieValues = [0.37, 0.4, 0.5, 0.35]
```

```python
# Plot a better chart using this data
plt.pie(badPieValues, labels=badPieNames)
plt.savefig("badPie.png")
```

## Bad Bar Chart 1

```python
badBarNames = ["A", "B", "C"]
badBarValues = [240, 242, 250]
```

```python
# Plot a better chart using this data
ind = np.arange(3)

plt.bar(ind, badBarValues, align="center")
plt.xticks(ind, badBarNames)
plt.ylim(239,252)
plt.savefig("badBar.png")
```
